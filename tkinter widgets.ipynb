```python
import tkinter
```

```python
window = tkinter.Tk()
window.minsize(width = 600, height = 500)

def Yadh():
    label = tkinter.Label()
    label.config(text = "I am Yadh")
    label.pack()

button = tkinter.Button(text = "Click me to become Yadh", command = Yadh)
button.pack()

# Textbox

text = tkinter.Text(width = 20, height = 5)
text.focus()
text.insert("1.0","Type Anything")
text.pack()

label2 = tkinter.Label()

def done():
    label2.config(text = "You finished typing")
    label2.pack()
    print(text.get("1.0", tkinter.END)) # 1.0 means text starting at first line at charcter 0, till'st the end.

button2 = tkinter.Button(text = "Done", command = done)
button2.pack()

#spinbox

label2 = tkinter.Label()
def spin():
    s = spinbox.get()
    label2.config(text = f"Spin chosen is = {s}")
    label2.pack()
    

spinbox = tkinter.Spinbox(from_ = 0, to = 10, width = 5, command = spin)
spinbox.pack()

label3 = tkinter.Label()

#scale does not have get() function. Instead, it directly uses passed value
def scal(value):
    #sc = scale.get()
    label3.config(text = f"Scale is: {value}")
    label3.pack()

scale = tkinter.Scale(from_ = 0, to = 100, command = scal)
scale.pack()

#checkbox

checked_state = tkinter.IntVar()
label4 = tkinter.Label()
def use():
    c = checked_state.get()
    label4.config( text = f"{c}")
    label4.pack()
    
    
checkbox = tkinter.Checkbutton(text = "Is on?", variable = checked_state, command = use)
checkbox.pack()

# radio
label5 = tkinter.Label()
radio_state = tkinter.IntVar()
def rad():
    r = radio_state.get()
    label5.config(text = f"r")
    label5.pack()

radio1 = tkinter.Radiobutton(text = "option 1", value = 1, variable = radio_state, command = rad)
radio2 = tkinter.Radiobutton(text = "option 2", value = 2, variable = radio_state, command = rad)
radio1.pack()
radio2.pack()

# Listbox

def l():
    print(listbox.get(listbox.curselection()))

listbox = tkinter.Listbox(height = 4)
fruits = ["apple", "banana", "pear", "mango"]
for f in fruits:
    listbox.insert(fruits.index(f), f)

listbox.bind("<<ListboxSelection>>", l)
listbox.pack()
window.mainloop()
```
