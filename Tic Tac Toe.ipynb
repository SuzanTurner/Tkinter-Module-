```python
import tkinter

window = tkinter.Tk()
window.minsize(height = 221, width = 247)
window.title("Tic Tac Toe")

b = True

def reset_game():
    global b
    for button in button_dict.values():
        button.config(text="", state="normal")
    game_over.grid_forget() 
    b = True
    
def play(n):
    global b
    if b == True:
        button_dict[f"button_{n}"].config(text="X", font=("Helvetica", 9, "bold"))
        b = -b
    else:
        button_dict[f"button_{n}"].config(text="O", font=("Helvetica", 9, "bold"))
        b =-b
    button_dict[f"button_{n}"].config(state="disabled")
    logic()

game_over = tkinter.Label()

def logic():
    win = False
    if button_dict["button_1"].cget("text") == button_dict["button_2"].cget("text") == button_dict["button_3"].cget("text"):
        win = True

    elif button_dict["button_4"].cget("text") == button_dict["button_5"].cget("text") == button_dict["button_6"].cget("text"):
        win = True
    
    elif button_dict["button_7"].cget("text") == button_dict["button_8"].cget("text") == button_dict["button_9"].cget("text"):
        win = True
        
    elif button_dict["button_1"].cget("text") == button_dict["button_4"].cget("text") == button_dict["button_7"].cget("text"):
        win = True
        
    elif button_dict["button_2"].cget("text") == button_dict["button_5"].cget("text") == button_dict["button_8"].cget("text"):
        win = True
        
    elif button_dict["button_3"].cget("text") == button_dict["button_6"].cget("text") == button_dict["button_9"].cget("text"):
        win = True
        
    elif button_dict["button_1"].cget("text") == button_dict["button_5"].cget("text") == button_dict["button_9"].cget("text"):
        win = True
        
    elif button_dict["button_3"].cget("text") == button_dict["button_5"].cget("text") == button_dict["button_7"].cget("text"):
        win = True

if win == True:
    game_over.config(text="Game Over", font=("Helvetica", 9, "bold"))
    game_over.grid(row=4, column=1, columnspan=3)
else:
    game_over.config(text="Draw", font=("Helvetica", 9, "bold"))
    game_over.grid(row=4, column=1, columnspan=3)
    

button_dict = {}

button_1 = tkinter.Button(command = lambda : play(1))
button_1.config(highlightthickness = 2, height = 4, width = 10)
button_1.grid(row = 1, column = 1)
button_dict["button_1"] = button_1

button_2 = tkinter.Button(command = lambda : play(2))
button_2.config(highlightthickness = 2, height = 4, width = 10)
button_2.grid(row = 1, column = 2)
button_dict["button_2"] = button_2

button_3 = tkinter.Button(command = lambda : play(3))
button_3.config(highlightthickness = 2, height = 4, width = 10)
button_3.grid(row = 1, column = 3)
button_dict["button_3"] = button_3

button_4 = tkinter.Button(command = lambda : play(4))
button_4.config(highlightthickness = 2, height = 4, width = 10)
button_4.grid(row = 2, column = 1)
button_dict["button_4"] = button_4

button_5 = tkinter.Button(command = lambda : play(5))
button_5.config(highlightthickness = 2, height = 4, width = 10)
button_5.grid(row = 2, column = 2)
button_dict["button_5"] = button_5

button_6 = tkinter.Button(command = lambda : play(6))
button_6.config(highlightthickness = 2, height = 4, width = 10)
button_6.grid(row = 2, column = 3)
button_dict["button_6"] = button_6

button_7 = tkinter.Button(command = lambda : play(7))
button_7.config(highlightthickness = 2, height = 4, width = 10)
button_7.grid(row = 3, column = 1)
button_dict["button_7"] = button_7

button_8 = tkinter.Button(command = lambda : play(8))
button_8.config(highlightthickness = 2, height = 4, width = 10)
button_8.grid(row = 3, column = 2)
button_dict["button_8"] = button_8

button_9 = tkinter.Button(command = lambda : play(9))
button_9.config(highlightthickness = 2, height = 4, width = 10)
button_9.grid(row = 3, column = 3)
button_dict["button_9"] = button_9

def disable_buttons():
    for button in button_dict.values():
        button.config(state="disabled")

reset_button = tkinter.Button(window, text="Restart", command=reset_game)
reset_button.grid(row=5, column=1, columnspan=3)

window.mainloop()
```
