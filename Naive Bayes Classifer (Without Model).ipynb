```python
print("Using Naive Bayes Classifier we shall predict if you can Play Golf based on certain factors")
outlook = input("Enter Outlook (Rainy, Overcast, Sunny):").lower()
temp = input("Enter Temperature (Hot, Mild, Cool):").lower()
hum = input("Enter Humidity (High, Normal):").lower()
wind = bool(input("Enter Wind (True, False):").title())
```

```python
def P_outlook_yes(outlook):
    if outlook == "sunny":
        p_outlook = ((df.Outlook == "Sunny") & (df["Play Golf"] == "Yes")).sum()/(df["Play Golf"] == "Yes").sum()
    elif outlook == "rainy":
        p_outlook = ((df.Outlook == "Rainy") & (df["Play Golf"] == "Yes")).sum()/(df["Play Golf"] == "Yes").sum()
    else:
        p_outlook = ((df.Outlook == "Overcast") & (df["Play Golf"] == "Yes")).sum()/(df["Play Golf"] == "Yes").sum()
    #print(p_outlook)
    return p_outlook

def P_outlook_no(outlook):
    if outlook == "sunny":
        n_outlook = ((df.Outlook == "Sunny") & (df["Play Golf"] == "No")).sum()/(df["Play Golf"] == "No").sum()
    elif outlook == "rainy":
        n_outlook = ((df.Outlook == "Rainy") & (df["Play Golf"] == "No")).sum()/(df["Play Golf"] == "No").sum()
    else:
        n_outlook = ((df.Outlook == "Overcast") & (df["Play Golf"] == "No")).sum()/(df["Play Golf"] == "No").sum()
    #print(n_outlook)
    return n_outlook

def P_temp_yes(temp):
    if temp == "hot":
        p_temp = ((df["Temperature"] == "Hot") & (df["Play Golf"] == "Yes")).sum()/(df["Play Golf"] == "Yes").sum()
    elif temp == "mild":
        p_temp = ((df["Temperature"] == "Mild") & (df["Play Golf"] == "Yes")).sum()/(df["Play Golf"] == "Yes").sum()
    else:
        p_temp = ((df["Temperature"] == "Cool") & (df["Play Golf"] == "Yes")).sum()/(df["Play Golf"] == "Yes").sum()
    #print(p_temp)
    return p_temp

def P_temp_no(temp):
    if temp == "hot":
        n_temp = ((df.Temperature == "Hot") & (df["Play Golf"] == "No")).sum()/(df["Play Golf"] == "No").sum()
    elif temp == "mild":
        n_temp = ((df.Temperature == "Mild") & (df["Play Golf"] == "No")).sum()/(df["Play Golf"] == "No").sum()
    else:
        n_temp = ((df.Temperature == "Cool") & (df["Play Golf"] == "No")).sum()/(df["Play Golf"] == "No").sum()
    #print(n_temp)
    return n_temp

def P_hum_yes(hum):
    if hum == "high":
        p_hum = ((df.Humidity == "High") & (df["Play Golf"] == "Yes")).sum()/(df["Play Golf"] == "Yes").sum()
    else:
        p_hum = ((df.Humidity == "Normal") & (df["Play Golf"] == "Yes")).sum()/(df["Play Golf"] == "Yes").sum()
    #print(p_hum)
    return p_hum

def P_hum_no(hum):
    if hum == "high":
        n_hum = ((df.Humidity == "High") & (df["Play Golf"] == "No")).sum()/(df["Play Golf"] == "No").sum()
    else:
        n_hum = ((df.Humidity == "Normal") & (df["Play Golf"] == "No")).sum()/(df["Play Golf"] == "No").sum()
    #print(n_hum)
    return n_hum

def P_wind_yes(wind):
    if wind == True:
        p_wind = ((df.Windy == True) & (df["Play Golf"] == "Yes")).sum()/(df["Play Golf"] == "Yes").sum()
    else:
        p_wind = ((df.Windy == False) & (df["Play Golf"] == "Yes")).sum()/(df["Play Golf"] == "Yes").sum()
    #print(p_wind)
    return p_wind

def P_wind_no(wind):
    if hum == True:
        n_wind = ((df.Windy == True) & (df["Play Golf"] == "No")).sum()/(df["Play Golf"] == "No").sum()
    else:
        n_wind = ((df.Windy == False) & (df["Play Golf"] == "No")).sum()/(df["Play Golf"] == "No").sum()
    #print(n_wind)
    return n_wind

def play_golf(outlook, temp, hum, wind):
    yes = (df["Play Golf"] == "Yes").sum()/len(df)
    no = (df["Play Golf"] == "No").sum()/len(df)
    yes_golf = P_outlook_yes(outlook) * P_temp_yes(temp) * P_hum_yes(hum) * P_wind_yes(wind) * yes
    no_golf = P_outlook_no(outlook) * P_temp_no(temp) * P_hum_no(hum) * P_wind_no(wind) * no
    if (yes_golf > no_golf):
        print("Yes, we will Play Golf")
        #print(yes_golf)
    else:
        print("No, we will not Play Golf")
        #print(no_golf)

play_golf(outlook, temp, hum, wind)

```
